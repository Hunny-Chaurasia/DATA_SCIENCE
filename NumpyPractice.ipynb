1. Array Creation Methods

*np.array(): Converts standard Python iterables (like lists or tuples) into an N-dimensional array (ndarray).*

```python
arr1=np.array((1,2,3,4,5))
arr2=np.array([1,2,3,4,5])
print(arr1 ,"\n", arr2)
```

*np.arange(start, stop, step): Generates sequential numerical values over a specified interval (similar to Python's built-in range).*

```python
arr1= np.arange(9, 900, 10)
arr2= np.arange(9)
print(arr1 ,"\n\n\n", arr2)
```

*np.zeros(shape): Creates an array filled entirely with zeros of a specified size/dimension.*

```python
arr1=np.zeros((3,5))
arr2=np.zeros(3)
print(arr1 ,"\n\n\n", arr2)
```

*np.ones(shape): Creates an array filled entirely with ones.*

```python
arr1=np.ones((3,5))
arr2=np.ones(3)
print(arr1 ,"\n\n\n", arr2)
```

*np.linspace(start, stop, num): Generates evenly spaced numbers over a specified mathematical interval.*

```python
arr1=np.linspace(9, 900, 9)
arr2=np.linspace(9, 900, 9, dtype=int)
print(arr1 ,"\n\n\n", arr2)
```

*np.copy(array): Creates a completely independent copy of an existing array in memory (so modifying the copy won't accidentally change your original data).*


```python
arr2=np.copy(arr1)
print(arr2)
```

*np.full(shape, fill_value) : Fills an array entirely with a custom static number of your choice.*


```python
np.full((2, 2), 7)
```

*np.empty(shape) : Allocates memory for an array without initializing its values. It contains whatever random "garbage values" were left over in your computer's RAM.*

```python
np.empty((2,3))
```

```python
np.eye(3)
```

```python
import numpy as np

# Imagine this is a 3D dataset: e.g., 4 frames, 5 pixels high, 2 channels wide
original_data = np.random.rand(4, 5, 2)
print("Original Shape:", original_data.shape)
# Output: (4, 5, 2)

# --- Creating the Clones ---

# 1. A blank mask matching the exact size, filled with zeros
zero_mask = np.zeros_like(original_data)

# 2. A baseline multiplier matching the exact size, filled with ones
ones_multiplier = np.ones_like(original_data)

# 3. A bias matrix matching the exact size, filled with a specific value (e.g., -1)
bias_matrix = np.full_like(original_data, -1.0)

# Verification
print("Zero Clone Shape:", zero_mask)        # Output: (4, 5, 2)
print("Bias Clone Shape:", bias_matrix)      # Output: (4, 5, 2)
print("First element of bias:", ones_multiplier) # Output: -1.0
```

```python
import numpy as np

# Imagine this is a 3D dataset: e.g., 4 frames, 5 pixels high, 2 channels wide
original_data = np.random.rand(4, 5, 2)
print("Original Shape:", original_data.shape)
# Output: (4, 5, 2)

# --- Creating the Clones ---

# 1. A blank mask matching the exact size, filled with zeros
zero_mask = np.zeros_like(original_data)

# 2. A baseline multiplier matching the exact size, filled with ones
ones_multiplier = np.ones_like(original_data)

# 3. A bias matrix matching the exact size, filled with a specific value (e.g., -1)
bias_matrix = np.full_like(original_data, -1.0)

# Verification
print("Zero Clone Shape:", zero_mask.shape)        # Output: (4, 5, 2)
print("Bias Clone Shape:", bias_matrix.shape)      # Output: (4, 5, 2)
print("First element of bias:", bias_matrix[0,0,0]) # Output: -1.0
```

**3. Shape Manipulation Methods**

*.reshape(new_shape): Changes the structural dimensions of an array without changing its underlying data (e.g., turning a 1D array of 12 elements into a 2D $3 \times 4$ matrix).*

```python
flat_array = np.arange(12)
print(flat_array)
reshaped_array = flat_array.reshape((3, 4))
print(reshaped_array)
```

# Linear Algebra in Numpy

```python
A=np.array([[1,2,5],[2,4,6],[1,7,5]])
b=np.array([3,4,5])
```

**1. Matrix Inversion & Solving Systems**

*np.linalg.inv(A): Computes the multiplicative inverse of a square matrix ($A^{-1}$).*

```python
np.linalg.inv(A)
```

*np.linalg.solve(A, b): Computes the exact solution vector $x$ for the linear equation $Ax = b$ (more computationally stable and faster than manually doing inv(A) @ b).*

```python
np.linalg.solve(A, b)
```

*np.linalg.pinv(A): Computes the Moore-Penrose pseudo-inverse for non-square matrices.*

```python
np.linalg.pinv(A)
```
