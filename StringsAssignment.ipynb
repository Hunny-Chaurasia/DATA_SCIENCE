# Q.1 Write a program to accept a string and display the string with changed case(Change upper and lower case)

```python
#Method1
strInp=input("Enter String Value").swapcase()
print(strInp)
```

```python
#Method2
strInp=input("Enter String Value")
newStr=""
for i in strInp:
  if i.islower():
    newStr+=i.upper()
  elif i.isupper():
    newStr+=i.lower()
  else:
    newStr+=i
print(newStr)
```

# Q.6 Write a  program which accepts a strings and count the frequency of each vowel.

```python
strInp=input("Enter String Value..").lower()
freq_a=0
freq_e=0
freq_i=0
freq_o=0
freq_u=0
for i in strInp:
  if 'a' == i:
    freq_a+=1
  elif 'e' == i:
    freq_e+=1
  elif 'i' == i:
    freq_i+=1
  elif 'o' == i:
    freq_o+=1
  elif 'u' == i:
    freq_u+=1

print("Frequency of : \n a - ",freq_a,"\n e - ",freq_e,"\n i - ",freq_i,"\n o - ",freq_o,"\n u - ",freq_u)
```

# Q.9 Write a program that takes full name as input and displays the abbrevations of the first and middle names except the last name which is displayed as it is. Eg:
Robert Brett Roser output :R.B.Roser.

```python
Full_Name=input("Enter your Name e.g. Robert Brett Roser").split()

Abbr_Name=""
for i in range(0,len(Full_Name)):
  if i == len(Full_Name)-1:
    Abbr_Name+=Full_Name[i]
  else:
    Abbr_Name+=Full_Name[i][0]
    Abbr_Name+=". "
print("abbrevations of the first and middle names except the last name : " , Abbr_Name)
```

# Q.10 Write a Python program to get a string made of the first 2 and the last 2 chars from a given a string. If the string length is less than 2, return instead of the empty string.

Sample String : 'Deepanshu'

Expected Result : 'Dehu'

Sample String : 'dv'

Expected Result : 'dvdv'

Sample String : 'd'


```python
#Slicing Method
strInp=input("Enter String Value : ")
newStr=""
if len(strInp)<2:
  print("Empty string : ",newStr)
else:
  newStr+=strInp[0:2]
  newStr+=strInp[-2:]
print(newStr)
```

```python
#Indexing Method
strInp=input("Enter String Value : ")
newStr=""
if len(strInp)<2:
  print("Empty string : ",newStr)
else:
  for i in range(0,len(strInp)):
    if i < 2 or i > len(strInp)-3:
      newStr+=strInp[i]
print(newStr)
```

# Q.14 Write a Python program to remove the characters which have odd index values of a given string.
